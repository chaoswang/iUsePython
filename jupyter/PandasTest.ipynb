```python
import numpy as np
titanic_surivival=pandas.read_csv('titanic_train.csv')
titanic_surivival.head()
```

```python
age=titanic_surivival['Age']
#print(age.loc[0:10])
age_is_null=pandas.isnull(age)
# print(age_is_null)
age_null_true=age[age_is_null]
# print(age_null_true)
print(len(age_null_true))
good_ages=titanic_surivival['Age'][age_is_null==False]
correct_mean_age=sum(good_ages)/len(good_ages)
print(correct_mean_age)
```

```python
passenger_survival=titanic_surivival.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survival)
```

```python
port_stats=titanic_surivival.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.sum)
print(port_stats)
```
